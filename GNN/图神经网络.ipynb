### 1.1卷积

### 卷积数学定义

**卷积**是通过两个函数 $f$ 和 $g$ 生成第三个函数的数学运算，表征函数 $f$ 与经过翻转、平移后得到的函数 $g$ 的乘积函数围成的**曲边梯形的面积**。  

<div align=center><img src="pic/卷积.gif" width="350"/>连续函数傅里叶变换</div>

**连续函数卷积：设 $f(x)$、$g(x)$是 $R$ 上两个可积函数，作积分**:  

<center>$c(\tau)=f(x) * g(x) = \int^{\infty}_{-\infty}{f(\tau)g(t-\tau)}dt$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$(1)$</center> 

**离散序列卷积：设 $f(m)$、$g(n)$是长度 $N$ 的两个离散信号，作积分**:

<center>$c(n)=f(m) * g(n) = \sum\limits_{m=0}^{N-1}{f(m)g(n-m)}$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$(2)$</center>  

**例子**  
    
**掷骰子问题**

设两个筛子是 $f$ 与 $g$，$f(1)$ 表示筛子 $f$ 正面向上为1的概律，求解：同时掷两个筛子1次，它们正面朝上数字和为4的概率？
    
情况分析 
* $f$ 为1，$g$ 为3；  
* $f$ 为2，$g$ 为2；  
* $f$ 为3，$g$ 为1； 

三种情况的概率和即为答案，表达式为：

<center>$\sum\limits_{\tau=1}^{3}{f(\tau)g(4-\tau)}$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$(3)$</center>  

可视化表示：

<img src="pic/筛子案例.png" width="480"/> 


式 $(3)$ 就是序列 $[f(1),f(2),f(3)]$ 与序列 $[g(3),g(2),g(1)]$ 的卷积：

<center>$(f * g)[4]=\sum\limits_{\tau=1}^{3}{f(\tau)g(4-\tau)}$ &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$(4)$</center>  

**离散函数卷积：把一个离散函数翻转过来 $g(4-\tau)$，与另一个离散函数 $f(\tau)$ 在相应的定义域 $(1,2,3)$ 中，求内积。$\sum\limits_{\tau=1}^{3}{f(\tau)g(4-\tau)}=[f(1),f(2),f(3)]\cdot[g(3),g(2),g(1)]$**


### 卷积在CNN中的作用

* CNN中的卷积利用一个共享参数的过滤器，通过计算中心像素点以及相邻像素点的**加权和**构成特征图（feature map）实现**空间特征**的提取，其中加权系数就是CNN网路中需要学习的卷积参数；
* 在CNN网络训练中使用**随机初始化**卷积核权重参数，基于误差通过反向传播梯度下降进行迭代优化；

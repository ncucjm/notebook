# 操作
* is_empty():判断链表是否为空
* length():返回链表长度
* travel(): 遍历链表
* add(item):在链表的头部增加一个节点
* append(item):在链表的尾部增加一个节点
* insert(pos, item):在指定位置插入一个节点
* remove(item):删除节点
* search(item):查找节点是否存在

```python
class Node():
    def __init__(self, item=None):
        self.item = item
        self.pre = None
        self.next = None 
        
class BiLinkList():
    
    def __init__(self):
        self.head = None
    
    def is_empty(self):
        if self.head == None:
            return True
        else:
            return False
    
    def length(self):
        cur = self.head
        if cur == None:
            return cur
        count = 0
        while cur:
            cur = cur.next
            count += 1
        return count
    
     #头部添加节点
    def add(self,item):
        new_node = Node(item)
        
        #空双向列表
        if self.head == None:
            self.head = new_node
            return
        
        cur = self.head
        self.head = new_node
        new_node.next = cur
        cur.pre = new_node
            
    #尾部添加节点
    def append(self,item):
        new_node = Node(item)
        
        #空双向列表
        if self.head == None:
            self.head = new_node
            return
        
        cur = self.head
        
        while cur.next:
            cur = cur.next
            
        cur.next = new_node
        new_node.pre = cur
        
    
    def insert(self,item, pos=0):
        
        #在头部插入一个节点
        if pos == 0:
            self.add(item)
            return
        #在尾部插入一个节点
        if pos == self.length():
            self.append(item)
            return
        
        new_node = Node(item)
        
        #双向列表为空
        if self.head == None:
            self.head = new_node
            print('this is empty BiLinkList,the s% will be the first item',item)
            return
        
        #超出双向列表索引
        if pos > self.length():
            print('error out of the index')
            return
        
        #在双向列表中间插入一个节点
        cur = self.head
        
        while pos:
            pos = pos - 1
            cur = cur.next
            
        cur.pre.next = new_node
        new_node.next = cur
        
        
    def travel(self):
        
        if self.head == None:
            return self.head
        
        cur = self.head
        
        while cur:
            print(cur.item)
            cur = cur.next
            
    def remove(self,item):
        
        if self.head == None:
            print('this is empty BiLinkList')
            return
        
        cur = self.head
        find = None
        
        while cur and not find:
            
            if cur.item == item:
                #删除节点是双向链表的第一个节点
                if cur.pre == None:
                    self.head = cur.next
                else:
                    cur.pre.next = cur.next
                find = True
                
            cur = cur.next
        
        if find == None:
            print('no this item in the BiLinkList')
        else:
            print('finish')
            
    def search(self, item):
        
        if self.head == None:
            print('this is empty BiLinkList')
            return
        
        cur = self.head
        while cur:
            if cur.item == item:
                print('Ture')
                return
            cur = cur.next
        
        print('False')
        
if __name__=="__main__":
    B = BiLinkList()
    B.append(1)
    B.append(2)
    B.append(3)
    B.append(4)
    B.append(5)    
    B.search(6)
        
        
```
